```r
packageurl <- "https://cran.r-project.org/src/contrib/Archive/ouch/ouch_2.9-2.tar.gz"
install.packages(packageurl, repos=NULL, type="source")
library("ouch")
library("DescTools")
library("Biostrings")
library("seqinr")
library("stringr")
library("FNN")
library("ape")
library("RCurl")
library("diagram")
library("ggplot2")
library("corrplot")
library("qgraph")
library("circlize")
```

```r
#pdf("PLD_coevolution.pdf")
JTT_Gamma=read.csv("Coevolution_pairs_mammals/BGM_mammals_new_submission_for_plots.csv")

JTT_Gamma=JTT_Gamma[which(JTT_Gamma$Shared.subs > 1),]

colnames(JTT_Gamma)=c("Pair1", "Pair2", "Prob.", "Subs")
JTT_Gamma=as.data.frame(JTT_Gamma)

index_gamma=which(JTT_Gamma$Prob. > 0)

limit=525
BGM=matrix(0, nrow = length(index_gamma), ncol = 2)
BGM_num=matrix(0, nrow = length(index_gamma), ncol = 2)
BGM_num[,1]=JTT_Gamma$Pair1[index_gamma]
BGM_num[,2]=JTT_Gamma$Pair2[index_gamma]

for (i in 1:length(index_gamma)){
    
    if (JTT_Gamma$Pair1[index_gamma][i]<limit){
        BGM[i,1]="FUS"  #Region_call(JTT_Gamma$Pair1[index_gamma][i])     #"FUS"  
        
    } else{
        BGM[i,1]="TNPO1"
        BGM_num[i,1]=BGM_num[i,1]
    }
    
    
    if (JTT_Gamma$Pair2[index_gamma][i]<limit){
        BGM[i,2]="FUS"  #Region_call(JTT_Gamma$Pair2[index_gamma][i]) 
    } else{
        BGM[i,2]="TNPO1"
        BGM_num[i,2]=BGM_num[i,2]
    }
          
}

RED_spec=rgb((0:15)/15, green = 0, blue = 0, names = paste("red", 0:15, sep = "."))

par(mar = c(1, 1, 1, 1) ) 
circos.initialize(factors = rep("FUS", (161-32+1)), x = seq(32,161))

circos.trackPlotRegion(factors =BGM[,1], x=BGM_num[,1], y=BGM_num[,2] , bg.col = c("gray", "black") , bg.border = NA, track.height=0.1)

circos.axis(major.tick.percentage = 1, minor.ticks = 1)
    # # Add a link between a point and another

    for (i in 1:length(index_gamma)){
        
        if ( JTT_Gamma$Prob.[index_gamma][i] < 0.5){
        circos.link(BGM[i,1], BGM_num[i,1], BGM[i,2], BGM_num[i,2], h = 5, col = "gray", lwd=JTT_Gamma$Prob.[index_gamma][i]*5)
            } else {
        circos.link(BGM[i,1], BGM_num[i,1], BGM[i,2], BGM_num[i,2], h = 5, col = "red", lwd=JTT_Gamma$Prob.[index_gamma][i]*5) 
        }
    }


#dev.off()
```
